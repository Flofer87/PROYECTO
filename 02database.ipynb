###La imagen previa muestra un ejemplo de la información contenida en la base de datos, la que se desglosa de la siguiente forma:


1.   Año: indica el período académico donde se aplicó la encuesta docente.
2.   Docente Rut: indica el rut de cada profesor que obtuvo una encuesta docente válida.
3.   Docente Nombre: nombre completo de cada docente.
4.   Docente Sexo: sexo declarado de cada docente.
5.   Sem. 1: indica el resultado de evaluación docente obtenido en el semestre 1 del año previamente indicado.
6.   Asig. Sem 1: muestra la cantidad de asignaturas impartidas en el semestre previamente indicado.
7.   Sem 1 ID: indica el resultado obtenido en la evaluación docente del período académico indicado previamente. Este resultado se basa en la evaluación de los criterios antes mencionados (AVA, clima de aula, etc) y se refleja en forma numérica.
8.  Sem. 2: indica el resultado de evaluación docente obtenido en el semestre 2 del año previamente indicado.
9.  Asig. Sem 2: muestra la cantidad de asignaturas impartidas en el semestre previamente indicado.
10. Sem 2 ID: indica el resultado obtenido en la evaluación docente del período segundo semestre.
11. Total Asignaturas: muestra la totalidad de asignaturas impartidas por el docente en el año académico que correponde.
12. Encuestados: es el universo total de estudiantes habilitados para poder responder la encuesta docente, correpondientes a todas las secciones impartidas por el docente en el año académico indicado.
13. Respuestas válidas: corresponde al número total de estudiantes que respondieron la encuesta docente de forma efectiva, para que sea considerada como válida.
14. ID Global: indica el resultado final anual obtenido en la encuesta docente.
